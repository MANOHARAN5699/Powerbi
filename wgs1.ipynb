```python
# This converts many similar statuses into common dashboard categories
status_map = {

    # New leads
    "New": "New",

    # Leads that need follow-up
    "Followup": "Follow-up",
    "Close follow up": "Follow-up",
    "Call later": "Follow-up",
    "Busy": "Follow-up",
    "Proposal pending": "Follow-up",

    # Interested leads
    "Appointment": "Interested",
    "Appointment fixed": "Interested",
    "Proposal sent": "Interested",
    "Details has sent through whatsapp": "Interested",
    "On process": "Interested",

    # Leads not interested
    "Not interested": "Not Interested",
    "Least interested": "Not Interested",
    "Not required @ present": "Not Interested",

    # Leads not reachable
    "Not answering": "Unreachable",
    "Not reachable": "Unreachable",
    "Switched off": "Unreachable",
    "Contact attempted": "Unreachable",

    # Invalid leads
    "Wrong lead": "Invalid",
    "Wrong number": "Invalid",
    "Invalid": "Invalid",

    # Converted leads
    "Customer": "Converted",
    "Course": "Converted",

    # Lost leads
    "Lost": "Lost",
    "Closed by other team": "Lost",
    "Already have one": "Lost",

    # Other cases
    "Looking for job": "Other",
    "Repeated contact": "Other",
    "Other language": "Other"
}


```

```python
df['Status']=df['Status'].replace(status_map)
```

```python
df['Status'].unique()
```

```python
df["Source"] =df["Source"].replace('Social media dm','digital marketing')
```

```python
df['Source']=df['Source'].str.strip()
```
